```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
ds = pd.read_csv('Data_Set/AB_NYC_2019.csv')
ds
```

```python
sn.countplot(x='neighbourhood_group', data=ds , palette='Set2')
plt.show()
```

```python
sn.barplot(x='neighbourhood_group',y='price', data=ds, palette='Set2')
plt.show()
```

```python
# sn.barplot(x='neighbourhood_group', y='price', data=ds,ci = False, palette='Set2') # ci = False means that the confidence interval is not drawn.
sn.barplot(x='neighbourhood_group', y='price', data=ds,errorbar=('ci', False) , palette='Set2')
plt.show()
```

```python
plt.hist(ds['price'])
plt.xlabel('Price(in us dollars $)')
plt.show()
```

```python
plt.hist(ds['price'] , bins = np.arange(0, 1000, 50))
plt.xlabel('Price(in us dollars $)')
plt.show()

```

```python
plt.scatter(x = ds['price'], y = ds['number_of_reviews'])
plt.xlabel('Price(in us dollars $)') #  x-axis label
plt.ylabel('number_of_reviews') #  y-axis label
plt.show()
```

```python
plt.scatter(x = ds['price'] , y = ds['number_of_reviews'] )
plt.xlabel('Price(in us dollars $)') #  x-axis label
plt.ylabel('number_of_reviews') #  y-axis label
plt.xlim(0, 1000)
plt.show()
```

```python
plt.scatter(x = ds['price'] , y = ds['number_of_reviews'] , s= 5)
plt.xlabel('Price(in us dollars $)') #  x-axis label
plt.ylabel('number_of_reviews') #  y-axis label
plt.xlim(0, 1000)
plt.show()
```
